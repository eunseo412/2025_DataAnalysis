```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import platform
```

```python
# seaborn에 있는 sample data
iris_data = sns.load_dataset('iris')
```

```python
plt.boxplot(iris_data['sepal_width'])
```

```python
# matplot에 범례 추가: legend()
# 보조선 추가: grid()
# 그래프 제목: title()
# x축, y축의 이름을 지정: xlabel(), ylabel()
x = np.linspace(0, 10, 200)
x
```

```python
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y1, label = 'sin')
plt.plot(x, y2, label = 'cos')

# 그래프에 격자 추가
plt.grid(True, axis = 'both')
# x축과 y축에 이름을 생성
plt.xlabel('X')
plt.ylabel('Y')
# 그래프 제목
plt.title('sin, cos graph')

plt.legend(loc = 'upper left')
# plt.legend(['Sin', 'Cos'])
# plt.legend()

plt.show()
```

```python
if platform.system() == 'Darwin':
    plt.rc('font', family = 'AppleGothic')
else:
    plt.rc('font', family = 'MalgunGothic')
```

```python
x = ['국어', '영어', '수학', '과학']
scores = [80, 60, 70, 50]

plt.bar(x, scores)

# x축의 데이터들을 회전 -> x축 데이터에 접근하는 함수
plt.xticks(rotation = 45)
# 그래프에 문자를 추가 (주석과는 다른 함수)
# text(x축, y축, 문자)
# plt.text(0, 80, '80')

for i in range(len(scores)):
    plt.text(i, scores[i]+1, scores[i])
```

```python
drinks = pd.read_csv('../data_git/csv/drinks.csv')
drinks.info()
```

```python
corr = drinks.loc[:, 'beer_servings' : 'total_litres_of_pure_alcohol'].corr()
```

```python
corr
```

```python
plt.figure(figsize = (8, 5))
sns.heatmap(
    data = corr,
    annot = True, # 주석 표시 유무
    square = True, # 히트맵 정사각형 표시 유무
    fmt = '0.3f',    # 주석의 소수점 자리수 표시
    annot_kws = {'size' : 20},   # 주석의 환경 설정 폰트 크기 지정
    cbar = False,    # 색상 바를 표시 유무
    xticklabels = ['beer', 'spirit', 'wine', 'alcohol'],
    yticklabels = ['beer', 'spirit', 'wine', 'alcohol']
)
plt.show()
```

```python
sns.pairplot(
    data = drinks.loc[:,
                      'beer_servings' : \
                        'total_litres_of_pure_alcohol'],
    kind = 'reg'
)
plt.show()
```

```python
# subplots() : 그래프가 그려지는 공간을 나눠서 여러 개의 그래프를 그려주는 함수
# plt.subplot(2, 2, 1)
fig, axes = plt.subplots(2, 2)
# axes[0][0].plot([1, 2, 3, 4])
# axes[1][1].plot([4, 3, 2, 1])

axes.flat[1].plot([1, 2, 3, 4])

plt.show()
```

```python
cols = ['beer_servings', 'spirit_servings', 'wine_servings', 'total_litres_of_pure_alcohol']
fig, axes = plt.subplots(3, 1, figsize = (5, 15))

for i in range(1, len(cols)):
    ax = axes.flat[i-1]
    ax.scatter(drinks[cols[0]], drinks[cols[i]])

plt.show()
```

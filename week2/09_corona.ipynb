```python
# 인덱스를 초기화
# 초기화(reset) + 위치(index)
# reset_index() -> 인덱스를 재지정하고 기존의 인덱스는 컬럼의 데이터로 변환
# 매개변수 inplace -> 원본의 데이터를 수정할것인가?
# 매개변수 drop -> 기존의 인덱스를 제거할 것인가?
df.reset_index(drop = True, inplace = True)

## drop 을 하지 않는 이상 데이터가 삭제되는 일은 없다.
```

```python
# 총확진자수 데이터를 이용해서 일일확진자 데이터를 생성
# 기존의 데이터를 이용하여 새로운 데이터를 생성 -> 파생변수
# 기존의 데이터만으로 설명이 불가능하거나 아니면 상관 관계가 약한 경우
# 새로운 데이터를 생성해서 설명이 가능하고 상관관계를 올릴 수 있도록 하는 방법

# 일일 확진자 -> 오늘의 총확진자수 - 어제의 총확진자수
df.tail()
```

```python
decide_list = [0]
for i in range(1, 820):
    decide = int(df.loc[i, '총확진자'] - df.loc[i-1, '총확진자'])
    decide_list.append(decide)

decide_list
```

```python
# dict에서 key, value 추가하는 형태와 같다.
# value가 series 형태인게 dataframe

df['일일확진자'] = decide_list
df.head()
```

```python
df['총사망자'].shift(-1)
# 결측치가 들어가는 데이터는 int 형태로 안들어가고 float으로 들어감
```

```python
df['총사망자'].values
```

```python
# index를 이동하는 함수
# shift(n) -> n만큼 인덱스를 변경. n의 기본값은 1
df['일일사망자'] = df['총사망자'] - df['총사망자'].shift()
df
```

```python
# 특정 시리즈 안에서의 차이를 구하는 함수
# diff(n) -> n: index의 차이값
df['총사망자'].diff()
```

```python
# 일일사망자 컬럼의 데이터에서 결측치는 0으로 채워준다.
df['일일사망자'].fillna(0, inplace = True)
```

# item_price 데이터에 결측치 채우기
- 상품별 결측치를 제외한 item_price의 평균을 구한다.
- 해당 상품의 item_price가 결측치인 데이터에 평균값을 대입

```python
# 상품A의 item_price가 결측치가 아닌 데이터를 확인

flag = uriage['item_name'] == '상품A'
flag_null = uriage['item_price'].isna()
```

```python
# item_name이 상품A이고 (그리고) item_price가 결측치가 아닌(~flag_null) 상품 정보(모든 컬럼->컬럼 조건식은 없음)를 확인
uriage.loc[
    flag & ~flag_null, 
]
# 차원유지. 조건이 단일 데이터가 아니기 때문. (flag랑 flag_null은 Series임)
# 컴퓨터는 type으로 단일인지 아닌지를 판단
```

```python
#위의 인덱스 조건에서 item_price 컬럼(컬럼의 조건식)의 데이터의 평균값 mean()을 구한다.
uriage.loc[
    flag & ~flag_null, 
    'item_price'
]
# 차원이 하나 낮아짐. 조건이 단일이기 때문에(str)
```

```python
#위의 인덱스 조건에서 item_price 컬럼(컬럼의 조건식)의 데이터의 평균값 mean()을 구한다.
uriage.loc[
    flag & ~flag_null, 
    ['item_price']
]
# 차원이 그대로 DataFrame 조건이 단일이 아니기 때문에(list)
```

```python
mean_price = uriage.loc[
    flag & ~flag_null, 
    'item_price'
].mean()
```

```python
len(uriage.loc[flag, 'item_price'])
# .fillna(mean_price, inplace = True)
```

```python
#uriage
```

```python
uriage.loc[flag & ~flag_null, 'item_price'] = uriage.loc[flag & ~flag_null, 'item_price'].fillna(mean_price)
```

```python
uriage
```

```python
# 위의 작업들을 상품A부터 상품Z까지(item_name 데이터에서 중복으로 제거한 리스트) 반복 실행
item_list = uriage['item_name'].unique()

for item_name in item_list:
    flag = uriage['item_name'] == item_name
    flag_null = uriage['item_price'].isna()

    mean_price = uriage.loc[flag & ~flag_null, 'item_price'].mean()
    uriage.loc[flag & flag_null, 'item_price'] = mean_price
```
